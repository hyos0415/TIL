```python
# python은 객체지향언어 입니다. 추가적으로 함수적 프로그래밍도 할 수 있어요!

# python에서 나오는 모든것들은 "객체(instance, object)"
# 숫자, 리스트, 튜플, range, map, set, bool

# 객체는 class로부터 파생되요!
# class안에는 상태(변수), 행위(함수)
# 객체도 변수(property, field, member variable), 함수(method)

class Student(object):
    
    # method(정해져있는 특별한 일을 하는 method)
    def __init__(self, name, dept):
        self.name = name       # self.name : property
        self.dept = dept
        
    # method    
    def get_student_info(self):
        print(self.name, self.dept)
   
    def set_stu_name(self, name):
        self.name = name
        
stu1 = Student('홍길동','CS')


print(stu1.name)         # property를 제어
stu1.name = '최길동'     # property를 변경 (객체지향 측면에서는 바람직 하지 않아요)
                        # 오류가 발생할 여지가 많기 때문에 직접적인 제어는 지양  
stu1.set_stu_name('최길동')

stu1.get_student_info()  # method를 호출
```

```python
# python은 객체지향면에서 잘 맞지 않는 부분들도 가지고 있어요!
class Student(object):
    
    # method(정해져있는 특별한 일을 하는 method)
    def __init__(self, name, dept):
        self.name = name       # self.name : property
        self.dept = dept
        
    # method    
    def get_student_info(self):
        print(self.name, self.dept)
   
    def set_stu_name(self, name):
        self.name = name
        
stu1 = Student('홍길동','CS')

# 당연히 class안에 존재하는 property와 method만 이용할 수 있어요!
stu1.address = '서울'   # 객체지향관점에서 봤을때는 당연히 오류. 허용하면 안되요!
                       # 프로그래밍의 유연성을 제공하기 위해 동적 property, 
                       # method를 허용
print(stu1.address)
```
